In this assignment you will create a `time` class, which will be useful for measuring lengths of time.

The attributes should be `hour`, `minute`, and `second`. You may assume the user has entered numbers between 0 and 59 for both the minute and second when they create a `time` object.

The `repr` function should display the time using the format hours:minutes:seconds. However, if the number of minutes or seconds is a single digit, then you should add a 0 in front.

You will create five methods using operator overloading, so that times can be compared with <, ==, and <=, added with +, and subtracted with -. Be careful! When adding and subtracting, the result should be a time with the number of minutes and seconds between 0 and 59. Study the examples below to make sure you understand how this should work.

```
> time1=time(1,2,3)
> time1.minute

> time1

> time2=time(1,59,58)
> time1<time2
True

> time1==time2

> time1<=time2
True

> time1+time2

````python
class time():
  def __init__(self,hour,minute,second):
    self.hour=hour
    self.minute=minute
    self.second=second

  def __repr__(self):
    hour=str(self.hour)
    min=str(self.minute)
    sec=str(self.second)
    if self.minute<9:
      min='0'+str(self.minute)
    if self.second<9:
      sec='0'+str(self.second)
    return hour+':'+min+':'+sec

  def __lt__(self,other):
    if self.hour<other.hour:
      return True
    elif self.minute<other.minute:
      return True
    elif self.second<other.second:
      return True
    else:
      return False

  def __eq__(self,other):
    if self.hour!=other.hour:
      return False
    elif self.minute!=other.minute:
      return False
    elif self.second!=other.second:
      return False
    else:
      return True

  def __le__(self,other):
    result=False
    if self.hour<other.hour or self.hour==other.hour:
      return True
    elif self.minute<other.minute or self.minute==other.minute:
      return True
    elif self.second<other.second or self.second==other.second:
      return True
    return result


  def __add__(self,other):
    H=int(self.hour+other.hour)
    M=int(self.minute+other.minute)
    S=int(self.second+other.second)
    if S>59:
      M+=1
      S=S%60
    if M>59:
      H+=1
      M=M%60
    return time(H,M,S)

  def __sub__(self,other):
    hour=self.hour-other.hour
    min=self.minute-other.hour
    sec=self.second-other.second
    if sec<0:
      min=int(min)-1
      sec=sec%60
    if min<0:
      hour=int(hour)-1
      sec=sec%60
    return time(hour,min,sec)
````
