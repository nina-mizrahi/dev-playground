In this problem you will implement a *bubble sort*. This will take a list and put it in increasing order. The algorithm is as follows:

1. Examine the first pair of numbers in the list. If the first is bigger than the second, then switch their order.
2. Repeat for the index 1 and 2 elements of the list.
3. Repeat for the index 2 and 3 elements, etc, until you get to the end of the list.
4. If you had switched the order of ANY elements in the list, then go back to step 1 and repeat. If not, return the final list.

*Hint.* To switch the order of elements i and i+1 in a list L, you can do this:

`L=L[:i]+[L[i+1]]+[L[i]]+L[i+2:]`


```
L=[4,5,2,7,1,9,3,5]
bubble(L)
> [1, 2, 3, 4, 5, 5, 7, 9]

W=['cherry','banana','apple']
bubble(W)
> ['apple', 'banana', 'cherry']
```

````python
def bubble(L):
  flag=0
  while flag==0:
    flag=1
    for i in range(len(L)-1):
      if L[i]>L[i+1]:
        L=L[:i]+[L[i+1]]+[L[i]]+L[i+2:]
        flag=0
  return L
````

````python
L=[4,5,2,7,1,9,3,5]
````

````python
bubble(L)
````

````python
W=['cherry','banana','apple']
bubble(W)
````
