In this assignment you will simulate the way a computer circuit carries out binary addition.

For the first function, you will return the `sum` bit from three-bit addition. There are four possible combinations of values for x, y, and z that should produce a `sum` bit of 1. Express each possibility with some combination of `and`, `or`, and `not` commands.

```
> bitsum(1,0,0)
1

> bitsum(1,1,0)
0

> bitsum(1,1,1)
1
```


````python
def bitsum(x,y,z):
  poss1=(not x) and (not y) and z
  poss2=x and (not y) and (not z)
  poss3=(not x) and y and (not z)
  poss4=x and y and z
  return int(poss1 or poss2 or poss3 or poss4)
````

In the second function, you will return the `carry` bit from three-bit addition. Again, there are four possible combinations should produce a `carry` bit of 1.

```
> bitcarry(1,0,1)
1

> bitcarry(0,1,0)
0

> bitcarry(1,1,1)
1
```

````python
def bitcarry(x,y,z):
  poss1=x and y and (not z)
  poss2=x and y and z
  poss3=(not x) and y and z
  poss4=x and (not y) and z
  return int(poss1 or poss2 or poss3 or poss4)
````

````python
bitcarry(1,1,1)
````

You are now ready to implement binary addition. Assume the inputs x and y to the `binarysum` function are four-element lists of 0's and 1's, representing some number in binary. Your function should return a five-elemen list, representing their sum.

```
> binarysum([1,1,1,0],[1,0,1,1])

[1,1,0,0,1]
```

````python
def binarysum(x,y):
  b0=bitsum(x[-1],y[-1],0)
  c0=bitcarry(x[-1],y[-1],0)
  b1=bitsum(x[-2],y[-2],c0)
  c1=bitcarry(x[-2],y[-2],c0)
  b2=bitsum(x[-3],y[-3],c1)
  c2=bitcarry(x[-3],y[-3],c1)
  b3=bitsum(x[-4],y[-4],c2)
  c3=bitcarry(x[-4],y[-4],c2)
  b4=c3
  return [b4,b3,b2,b1,b0]

````
