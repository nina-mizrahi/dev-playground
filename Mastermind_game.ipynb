Many years before there was Wordle, there was a similar board game called "Mastermind". Read the rules and history [here](https://en.wikipedia.org/wiki/Mastermind_(board_game)) and play it on-line [here](https://webgamesonline.com/mastermind/index.php).

In this assignment, you will program the game yourself. In our implementation, the computer will choose a "secret" list of four numbers from the digits 1 through 8 (no repeats). Then the player will make successive guesses what that list is.

Each time the player guesses, the computer will give feedback as to how close the guess is, in the form of a four character string with 'X's, 'O's, and '-'s. An 'X' represents one of the guessed numbers is both correct and in the correct position in the list. An 'O' means one of the guessed numbers is correct, but in the wrong position.

For example, if the secret code is 1234 and the player's guess is 1532, then the computer will respond with the string 'XXO-', because two of the guessed numbers are both correct and in the correct position (the 1 and the 3), and one of the numbers is correct, but in the wrong position (the 2).

To this end, you will need to write several helper functions. The first should take two lists of numbers (of the same length), and return the number of digits that match and are in the same position in both lists.

```
L=[2,3,5,7,10,0]
K=[2,3,0,5,10,9]

CorrectPositions(L,K)
> 3
```

Complete this function by filling in the missing list comprehension.

````python
def CorrectPositions(L,K):
  LC=[L[i]==K[i] for i in range(len(L))]
  return sum(LC)
````

````python
L=[2,3,5,7,10,0]
K=[2,3,0,5,10,9]

CorrectPositions(L,K)
````

Next, write a function `CorrectNumbers` which will take two lists of equal length, and return the number of elements of the first list that are also in the second (regardless of position). You may assume there are no repeated numbers in either list.

```
L=[2,3,5,7,10,0]
K=[2,3,0,5,10,9]

CorrectNumbers(L,K)
> 5
```

````python
def CorrectNumbers(L,K):
  LC=[1 for l in L if l in K]
  return sum(LC)
````

````python
CorrectNumbers(L,K)
````

Now use the previous two functions to write a function that takes two lists of equal length, and returns the number of digits that are in both lists, but not in the same positions.

```
L=[2,3,5,7,10,0]
K=[2,3,0,5,10,9]

WrongPosition(L,K)
> 2
```

````python
def WrongPosition(L,K):
  return CorrectNumbers(L,K)-CorrectPositions(L,K)
````

````python
WrongPosition(L,K)
````

Next, write a function `Score` that takes three numbers, n, m, and l, and outputs a string of total length l that starts with n 'X's, followed by m 'O's, and then finishes with '-'s.

```
Score(2,3,10)
>'XXOOO-----'
```

````python
def Score(n,m,l):
  return n*'X'+m*'O'+(l-m-n)*'-'
````

````python
Score(2,3,10)
````

Now we put it all together to evaluate a player's guess. Write a function `ScoreGuess(L,K)` that takes two lists and outputs a string with some number of 'X's for items in the same position in L and K, then some number of 'O's for items in both lists but in different positions, and then some number of '-'s. The total output string shoud have the same length as L.  

```
L=[2,3,5,7,10,0]
K=[2,3,0,5,10,9]

ScoreGuess(L,K)
> 'XXXOO-'
```

````python
def ScoreGuess(L,K):
  n=CorrectPositions(L,K)
  m=WrongPosition(L,K)
  l=len(L)
  return Score(n,m,l)
````

````python
ScoreGuess(L,K)
````
