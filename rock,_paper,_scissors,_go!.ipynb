In this problem you will create a function which takes two users "Rock, Paper, Scissors" choices, and return the winner.

````python
def RPSwinner(user1,user2):
  '''user1 and user2 are either 'rock', 'paper', or 'scissors'
     the function returns:
        0 if it is a tie
        1 if user1 wins
        2 if user2 wins
  '''
  x=user1
  y=user2
  if x=='rock':
    if y=='rock':
      return 0
    elif y=='paper':
      return 2
    elif y=='scissors':
      return 1
  elif x=='paper':
    if y=='rock':
      return 1
    elif y=='paper':
      return 0
    elif y=='scissors':
      return 2
  elif x=='scissors':
    if y=='rock':
      return 2
    if y=='paper':
      return 1
    if y=='scissors':
      return 0
````

To use your function to play Rock, Paper, Scissors, first run the following code block:

````python
def RPSgame():
  user1=input("What is player one's choice? ")
  user2=input("What is player two's choice? ")
  winner=RPSwinner(user1,user2)
  if winner==0:
    print("It's a tie!")
  else:
    print("Player ",winner," wins!")
  return
````

Then, run this to play with a friend:

````python
RPSgame()
````

Write a function ``Checkends(L)`` which takes in a string or a list, and returns True if the first letter/list entry is the same as the last, and False otherwise
```
CheckEnds([2,3,5,2])

> True

CheckEnds([2,3,7,9,10])

> False

CheckEnds('apple')

> False

CheckEnds('pop')

>True
```

````python
def CheckEnds(L):
  '''determines whether the first letter/entry of a list is the same as the last'''
  if L[0]==L[-1]:
    return True
  else:
    return False
````

````python
CheckEnds([2,3,5,2])
````

````python
CheckEnds([2,3,7,9,10])
````

````python
CheckEnds('apple')
````

````python
CheckEnds('pop')
````

Problem 2.

Write a function ``WordScramble(S)`` which takes a string and rearanges it so that every other letter, starting from the first, is at the beginning, and the remaining letters are at the end.

```
WordScramble('Computer')

> 'Cmueoptr'
```

````python
def WordScramble(S):
  '''rearanges a string so that every other letter, starting from the first, is at the beginning and the remaining letters are at the end'''
  return S[::2]+S[1::2]
````

````python
WordScramble('Computer')
````

Problem 3.

Write a function ``FlipSide(W)`` that takes a string W and flips the first half and the last half. If the length of the string is odd, then the middle letter shoud be included with the second half.


```
FlipSide('David')

>'vidDa'

FlipSide('matrix')

>'rixmat'
```

*Hint:* The length function will be useful here. If the length is odd, then you may also have to use the int() funtion after dividng by 2 to get a number that will be useful when slicing.

````python
def FlipSide(W):
  '''flips the first half and last half of a string'''
  x=int(len(W)/2)
  y=-int(len(W)-x)
  return W[y:]+W[0:x]
````

````python
FlipSide('David')
````

````python
FlipSide('matrix')
````
