In this homework you will write a `binary` class for storing binary numbers. Before the class definition, you wull need to write two functions that translate back and forth between binary and decimal.


The first function, `DecToBin` will take an integer x and return a list of 0's and 1's, representing the digits of the binary representation of x.

```
> DecToBin(10)
[1,0,1,0]

> DecToBin(15)
[1,1,1,1]
```

````python
def DecToBin(x):
  '''take an integer x and return a list of 0's and 1's, representing the digits of the binary representation of x.'''
  if x==1:
    return [1]
  else:
    if x/2==int(x/2):
      return DecToBin(x/2)+[0]
    else:
      return DecToBin((x-1)/2)+[1]

````

````python
DecToBin(10)
````

````python
DecToBin(15)
````

The second function, `BinToDec`, should do the exact opposite of `DecToBin`.

```
> BinToDec([1,0,1,0])
10

> BinToDec([1,1,1,1])
15
```


````python
def BinToDec(B):
  '''Takes a list B and returns the decimal value of that binary number'''
  if len(B)==1:
    return B[0]
  else:
    return 2*BinToDec(B[:-1])+B[-1]
````

````python
BinToDec([1,0,1,0])
````

````python
BinToDec([1,1,1,1])
````

We now use these to create a `binary` class for storing binary numbers. Your job is to complete this class by writing the repr function, so that binary numbers are displayed as a string of 0's and 1's (rather than a list).

```
> num=binary(10)
> num.bin
[1,0,1,0]

> num
1010

> num.decimal()
10
```

````python
class binary():
  def __init__(self,x):
    self.bin=DecToBin(x)

  def __repr__(self):
    result=''
    for i in self.bin:
      if i==1:
        result+='1'
      elif i==0:
        result+='0'
    return result


  def decimal(self):
    return BinToDec(self.bin)
````
