A famous unsolved question in Mathematics is called the *Collatz Conjecture*. Here's what it is about.

Suppose you start with a number n. If n is even, then replace it with n/2. If n is odd, replace it with 3n+1. Repeat these rules, but stop if you ever get to the number 1. Here's an example:
0. Start with n=3.
1. Since 3 is odd, we replace it with 3(3)+1=10.
2. Since 10 is even, we replace it with 10/2=5.
3. Since 5 is odd, we replace it with 3(5)+1=16.
4. Since 16 is even, replace it with 16/2=8.
5. Since 8 is even, replace it with 8/2=4.
6. Since 4 is even, replace it with 4/2=2.
7. Since 2 is even, replace it with 2/2=1.

The sequence we get is 3,10,5,16,8,4,2,1. The Collatz Conjecture states that **no matter what number you start with**, you will always end eventually up with 1.

As of 2020, the Collatz Conjecture was verified by computer for all starting numbers up to n=$2^{68}$. However, no one has provided a reason why it should be true for ALL possible starting numbers.

Write a function called `Collatz(n)` that starts with a number n, and returns how many steps it takes to get to the number 1 if you follow the rules described above. For example, `Collatz(3)` should return the number 7, because we have just seen that it takes 7 steps to get to the number 1. Here is some other data you can check:
```
Collatz(15)
> 17

Collatz(64)
> 6
```

````python
def Collatz(n):
  count=0
  while n>1:
    count+=1
    if n/2==int(n/2):
      n=n/2
    else:
      n=3*n+1
  return count
````

We continue our exploration of the Collatz Conjecture. You may have wondered: "If I keep checking numbers, how long would it take to find one that requires 10 steps? 100 steps? 1000 steps?"

Write a function `CollatzSteps(m)` that returns the first number that requires at least m Collatz steps. For example, `CollatzSteps(8)` should return the number 6, because:
Collatz(1)=0, Collatz(2)=1, Collatz(3)=7, Collatz(4)=2, Collatz(5)=5, and Collatz(6)=8. So 6 was the first number that requires 8 or more Collatz steps.

```
CollatzSteps(10)
> 7

CollatzSteps(100)
> 27

CollatzSteps(350)
> 77031
```

````python
def CollatzSteps(m):
  n=1
  while Collatz(n)<m:
    n+=1
  return n
````
